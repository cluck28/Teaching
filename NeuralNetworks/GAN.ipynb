<h1>Generative Neural Networks</h1>

GANs are an architecture designed to teach a neural network to create its own images or its own text. GANs are made up of two competing networks -- the generator and the discriminator -- that are trained through "adversarial learning". The generator tries to make up fake examples to fool the discriminator. While the discriminator tries to identify fake examples.

```python
from IPython.display import Image
Image(filename = "GAN.png", width=400, height=400)
```

The generator will start off making really obviously fake examples. But over time it will get better. As the generator gets better at generating fakes, the discriminator will also get better at telling real examples from fake examples. They eventually come to an equilibrium when the discriminator can no longer tell the difference between real and fake exmaples.
