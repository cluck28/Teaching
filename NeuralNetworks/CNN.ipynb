<h1>Convolutional Neural Networks</h1>

One area where neural networks have far outperformed other algorithms has been in the domain of image classification. The big advance that led to this success was in the architecture of convolutional neural networks. The basic idea of a convolutional neural network is to mimic how humans might observe an image. They try to identify low level features of an image and then combine these low level features into higher level features that might be correlated with the likelihood an image contains a specific object.

```python
from IPython.display import Image
Image(filename = "CNN.png", width=400, height=400)
```

The CNN accomplishes this with two types of layers: convolution layers (to find the features) and pooling layers (to combine the features).

<h2>Convolution Layers</h2>
Convolution layers can be thought of like a reference feature that is used to see if that feature apears in the original image. They are basically repeated multiplications of a sub matrix with the original image. If the region of the original image matches the convolution it is highlighted. It's kind of like taking a microscope that only knows how to spot diagonal lines and scanning it across an image. When the image has diagonal lines in it, the microscope picks them out.

<h2>Pooling Layers</h2>
Pooling layers are where CNNs combine the outputs of convolutions and aggregate them to give more macroscopic features. One such pooling operation is max pooling where a sub region is summarized just by taking the maximum value within that sub region:

```python
from IPython.display import Image
Image(filename = "poolingmath.png", width=400, height=400)
```

<h2>CNN Architecture</h2>
A CNN is then composed of a series of convolution layers followed by pooling layers in a repeated way.
