<h1>Recurrent Neural Networks</h1>

When the response we are trying to predict comes in a sequence we will want to use a specific architecture to make good predictions.

What counts as sequential data?

- words occuring in a sentence within a document
- time series of temperature, weather, etc
- financial time series
- speech or music
- handwriting

The basic idea is that RNNs find structure in the sequence of occurrences whereas CNNs were finding structure in the spatial order of things.

RNNs are often said to have memory -- they keep track of the previous state or states. Just like CNNS, RNNS have different architectures to handle different problems.

<h2>LSTM</h2>

Use gates to determine which previous information to save and use to make a prediction. It tries to forget irrelevant information to the prediction. Each RNN node is replaced by a "memory cell" that contains an internal state. This allows recent nodes to "forget" the irrelevant informatoin from the past.
